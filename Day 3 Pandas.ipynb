```python
tips = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
tips
```

```python
# Group by and aggregating
grouped = tips.groupby("sex")
```

```python
#will only work on numeric type columns
grouped.mean('tip','size')
```

```python
tips.groupby('sex').count()
```

```python
grouped.min('total_bill')
```

```python
tips.groupby('sex').agg({'total_bill':['mean','max','count','sum'], 'tip':['mean','max','count','sum']})
```
